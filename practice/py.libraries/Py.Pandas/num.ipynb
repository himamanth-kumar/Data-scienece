```python
import os
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
file_path = "./Q6-5-I B.Tech. I Semester (R22) Regular.csv"

# Load the CSV into a dataframe
df = pd.read_csv(file_path)

# Display the column names to verify them
print("Columns in the dataframe:", df.columns)

# Check if 'HTNO' exists in the dataframe
if "HTNO" in df.columns:
    # Define the roll number range (Example: '23Q61A0561' to '23Q61A05C0')
    start_roll = "23Q61A0561"
    end_roll = "23Q61A05C0"
    
    # Filter based on the roll number range (assuming HTNO is stored as strings)
    filtered_df = df[(df["HTNO"] >= start_roll) & (df["HTNO"] <= end_roll)]
    
    # Select only required columns (HTNO, SUBJECT_NAME, INTERNALMARKS, EXTERNALMARKS, TOTALMARKS, GRADE)
    required_columns = ["HTNO", "SUBJECT_NAME",  "EXTERNALMARKS", "GRADE"]
    filtered_df = filtered_df[required_columns]
    
    # Display the filtered results
    print(filtered_df)
    
    # Plot the table using Matplotlib
    fig, ax = plt.subplots(figsize=(10, 4))  # You can adjust the size as needed
    ax.axis('tight')
    ax.axis('off')
    
    # Create the table plot
    table = ax.table(cellText=filtered_df.values, colLabels=filtered_df.columns, loc='center', cellLoc='center')

    # Show the plot
    plt.show()
    
else:
    print("Column 'HTNO' not found!")
```
